#### Adding A Simple Recommendation Logic

```python
def recommend_netflix_content(title, num_recommendations=5):
    # 1. Find the details of the movie the user likes
    try:
        target_content = df_netflix[df_netflix['title'].str.lower() == title.lower()].iloc[0]
    except IndexError:
        return "Movie/TV Show not found. Please check the spelling!"

    # 2. Get the Genre and Rating of that movie
    target_genre = target_content['listed_in']
    target_rating = target_content['rating']

    # 3. Filter the dataframe for content with the same genre OR rating
    recommendations = df_netflix[
        (df_netflix['title'].str.lower() != title.lower()) & 
        ((df_netflix['listed_in'] == target_genre) | (df_netflix['rating'] == target_rating))
    ]

    # 4. Sort by release year to show the newest stuff first
    return recommendations[['title', 'type', 'release_year', 'rating', 'listed_in']].head(num_recommendations)

# Example Usage:
recommend_netflix_content('Blood & Water')
```

#### Netflix Analysis Dashboard

```python
# Set the overall style and Netflix-themed color palette
sns.set_style("whitegrid")
netflix_colors = ['#E50914', '#221f1f', '#564d4d', '#b3b3b3']

# Initialize a 2x2 grid for the dashboard
fig, axes = plt.subplots(2, 2, figsize=(20, 15))
fig.suptitle('Netflix Content Analysis Executive Dashboard', fontsize=30, fontweight='bold', color='#E50914')

# --- Plot 1: Content Type Split (Top Left) ---
type_counts = df_netflix['type'].value_counts()
axes[0, 0].pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', 
               startangle=140, colors=[netflix_colors[0], netflix_colors[1]], explode=(0.05, 0))
axes[0, 0].set_title('Movies vs TV Shows Distribution', fontsize=15)

# --- Plot 2: Content Growth Over Time (Top Right) ---
# Ensure date_added is datetime
df_netflix['year_added'] = df_netflix['date_added'].dt.year
growth = df_netflix.groupby(['year_added', 'type']).size().unstack().fillna(0)
sns.lineplot(data=growth, ax=axes[0, 1], palette=[netflix_colors[0], netflix_colors[1]], linewidth=3)
axes[0, 1].set_title('Content Addition Trends (By Year)', fontsize=15)
axes[0, 1].set_ylabel('Count of Titles')

# --- Plot 3: Top 10 Countries (Bottom Left) ---
top_countries = df_netflix['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, ax=axes[1, 0], 
            hue=top_countries.index, palette='Reds_r', legend=False)
axes[1, 0].set_title('Top 10 Content Producing Countries', fontsize=15)
axes[1, 0].set_xlabel('Total Titles')

# --- Plot 4: Rating Distribution (Bottom Right) ---
top_ratings = df_netflix['rating'].value_counts().head(10)
sns.barplot(x=top_ratings.index, y=top_ratings.values, ax=axes[1, 1], 
            hue=top_ratings.index, palette='viridis', legend=False)
axes[1, 1].set_title('Top 10 Content Ratings', fontsize=15)
plt.xticks(rotation=45)

# Adjust layout so labels don't overlap
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
